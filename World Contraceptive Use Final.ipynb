# Machine Learning Models
## Can you predict the year data was collected based on contraceptive method popularity?
From the preliminary analysis done above, there is a difference in preferred contraceptive method in the 1900s vs 2000s. To test this preliminary result, we will develop a machine learning model that uses data classified as either 0 for collection during the 1900s or 1 for collection during 2000s. This will be compared to the rates of all the different contraceptive method options in the survey. The data will be split into train and test, with test containing 75% of the data and train containing 25% of the data. k-NN, SVM, and decision tree will all be used to determine the best model. 

```python
#Specify x and y values, grab columns used for analysis

#make dataframe that grabs all of the different contraceptive method types
X=method.iloc[:,8:22]
#make specific variable that contains all of the collection years
year = method.iloc[:,2].values

#make function that will classify year as 0 or 1
y = [] 
def y_array(x):
    for i in x:
        if i >1999:
            y.append(1)
        if i<= 1999:
            y.append(0)
    return np.array(y)

y_array(year)
```

```python
#split the data into training and test
X_train, X_test, y_train, y_test= train_test_split(X, y, random_state=1, test_size=0.75)
```

## Classification using k-NN
First we are going to find the best k-value for the k-NN model, we will compare different k-values on the testing and training data and pick the best value overall. The results will be plotted help find the best k-value.

```python
#use cross val to find best k value

for k in [ 2, 3, 4, 5, 7, 10, 15, 20, 30]:
    print("k is:", k)
    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(X_train, y_train)

    y_pred = clf.predict(X_test)
    print("Accuracy Train:", metrics.accuracy_score(y_true=y_train, y_pred=clf.predict(X_train)))
    print("Accuracy Test:", metrics.accuracy_score(y_true=y_test, y_pred=y_pred))
 
```

```python
#Plot the k-values on the testing data
plt.figure(figsize=(16,9))
k = [ 2, 3, 4, 5, 7, 10, 15, 20, 30]
scores=[0.7031093279839519, 0.720160481444333, 0.7041123370110332, 0.7111334002006018,0.7141424272818455,0.6980942828485457,0.7121364092276831,0.7001003009027081,0.6880641925777332]
plt.plot(k, scores, 'ko', color = "red")
plt.xlim([0, 35])
plt.title("K-values vs Accuracy Score")
plt.ylim([0.6 , 0.8])
plt.grid()
plt.show()
```

## k-value Results
Based on the visualization we can see that the best k-value for the test data falls between 2-5. Specifically, the k-value that gives us the most accuracy on the testing data is 3. This value will be used in the k-NN model.

```python
#use best k value from cross-validation
#Create KNN classifier
knn = KNeighborsClassifier(n_neighbors = 3)
#Fit the classifier to the data
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
y_pred_train = knn.predict(X_train)

print('Accuracy on training data = ', metrics.accuracy_score(y_true = y_train, y_pred = y_pred_train))
print('Accuracy on test data = ', metrics.accuracy_score(y_true = y_test, y_pred = y_pred))

```

## k-NN Model Results
For the k-NN method the highest accuracy on the testing data is .720, this is relatively low for our model. To further test this model, we will try running it using SVM and decision tree to see if we can increase the accuracy on the test data. 

## Classification using SVM
SVM model will be used, first we will run cross validation to find the best fold accuracy and the best C value. 

```python
#calculate for accuracy of C values 2-20
c_accuracy = []
accuracy = []
for i in range(2,20):
    svclassifier = svm.SVC(kernel='rbf', C = i,gamma='scale')
    scores = cross_val_score(estimator = svclassifier, X = X_train, y = y_train, cv=9, scoring='accuracy')
    accuracy.append(scores.mean())
    c_accuracy.append((i,scores.mean()))
    
print(c_accuracy)
plt.figure(figsize=(16,9))
plt.xlim([0, 25])
plt.title("C paramter accuracy on testing data")
plt.plot(accuracy);
    
```

```python
#use best C value from cross-validation below
svclassifier = svm.SVC(kernel='rbf', C =9)
#fit the model to the training set
svclassifier.fit(X_train, y_train)
#use the test dataset to evaluate the model 
y_pred = svclassifier.predict(X_test)

# predict with the training data
y_pred_train = svclassifier.predict(X_train)
# measure accuracy
print('Accuracy on training data = ', metrics.accuracy_score(y_true = y_train, y_pred = y_pred_train))

# predict with the test data
y_pred = svclassifier.predict(X_test)
# measure accuracy
print('Accuracy on test data = ', metrics.accuracy_score(y_true = y_test, y_pred = y_pred))

```

## SVM Model Results
The SVM model on the testing data had an accuracy of .75 when compared to the k-NN model at. 72, it is slightly higher, meaning SVM may be a better model for the data. 

## Classification using Decison Tree
The last model type we are going to try is a decision tree. First, we will run cross validation to find the best parameters for min_samples_split and max depth accuracy.


```python
#use cross validation to calculate max depth accuracy 
depth = []
for i in range(3,20):
    clf = tree.DecisionTreeClassifier(max_depth=i)
    # Perform 7-fold cross validation 
    scores = cross_val_score(estimator=clf, X=X_train, y=y_train, cv=7, n_jobs=4)
    depth.append((i,scores.mean()))
print(depth)
```

## Cross Validation Results
Based on the results of this cross validation the best max depth for the decison tree model on the training data would be 4.

```python
#use cross validation to calculate min_samples_split
min_split = []
for i in range(2,30):
    clf = tree.DecisionTreeClassifier(min_samples_split=i)
    # Perform 7-fold cross validation 
    scores = cross_val_score(estimator=clf, X=X_train, y=y_train, cv=7, n_jobs=4)
    min_split.append((i,scores.mean()))
print(min_split)
```

## Cross Validation Results
The best minimum sample split for the model on the training data is 27.

```python
#Use parameters from the cross-validation below
decisionTree = tree.DecisionTreeClassifier(max_depth=4, min_samples_split=27)
decisionTree = decisionTree.fit(X_train, y_train)


# predict with the training data
y_pred_train = decisionTree.predict(X_train)
# measure accuracy
print('Accuracy on training data = ', metrics.accuracy_score(y_true = y_train, y_pred = y_pred_train))

# predict with the test data
y_pred = decisionTree.predict(X_test)
# measure accuracy
print('Accuracy on test data = ', metrics.accuracy_score(y_true = y_test, y_pred = y_pred))
```

## Decision Tree Results
The final model accuracy of the decision tree on the testing data is .76. Of the three different models, the decision tree model is the best fit for the data. An accuracy score of 76% on our model, indicates a good model overall. Thus, we can conclude that about 76% of the time our model can use rates of preferred contraceptive method to determine what year the data was collected.
