### Highest Used Contraceptive by Country Analysis

#### Can we determine the most used contraceptive by country? 

#### In trying to determine the most used contraceptive by country we decided we wanted to use the most recent data for a practical analysis to compare to present day. This data was filtered to display only 2018 to present, and the percentages of the contraceptive use shown by type. Next we grouped by the country to analyze the data up to this point. We dropped the countries Austria and Samoa because they were the only two countries with zero percentages for every contraceptive type. Next we reset the index, and renamed the column "Country or Area" to "Country" to simplify. 

```python
#filtering the data to end at 2018 to present.
contraceptives = method.loc[method["Survey\nend year"] > 2018]
#groupby the country to see percentages of types of contraceptive
contraceptives_grouped = contraceptives.groupby(by = ['Country or area'])['Female\nsterilization', 'Male\nsterilization', 'IUD', 'Implant', 'Injectable', 
                                   'Pill', 'Male condom', 'Female condom', 'Vaginal barrier methods', 'Lactational amenorrhea method (LAM)'].max()
#dropping the rows Austria and Samoa as they had zero percents across the rows. 
contraceptives_2 = contraceptives_grouped.drop(labels=["Austria", "Samoa"], axis=0)
#reset the index
contraceptives_index = contraceptives_2.reset_index()
#renaming the column Country for simplifying
contraceptives_new = contraceptives_index.rename({'Country or area':'Country'}, axis=1)
contraceptives_new
```

#### After filtering the data, we wanted to determine the most used contracpetive type. We did this by displaying the max percentage of each type to see the most used. From here we were able to determine IUD and the pill, are the most used types. We will proceed with an analysis of these two types., 

```python
#determining the most used contraceptive type
max = contraceptives_2.max(axis = 0)
print(max)
```

#### Although we ran a scatter plot and could see there is no linear relationship, we wanted to continue with the multilinear regression model to see the statistical analysis behind there being no relationship. ## Preliminary k-NN model
### Answering the question, can you predict if the year is the 1900s or 2000s based on popularity of contraceptive method?

```python
#grab all columns with contraceptive method types
X=method.iloc[:,8:22].values
#make a list of the survey years
year = method.iloc[:,2].values
#make a function that assigns 0 if the year is 1999 or below and 1 if it is above 1999
y = [] 
def y_array(x):
    for i in x:
        if i >1999:
            y.append(1)
        if i<= 1999:
            y.append(0)
    return np.array(y)

y_array(year)
```

```python
#split the data into test and training
X_train, X_test, y_train, y_test= train_test_split(X, y, random_state=1, test_size=0.75)
#use best k value from cross-validation that will be done as next steps
#Create KNN classifier
knn = KNeighborsClassifier(n_neighbors = 3)
#Fit the classifier to the data
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
y_pred_train = knn.predict(X_train)

print('Accuracy on training data = ', metrics.accuracy_score(y_true = y_train, y_pred = y_pred_train))

# measure accuracy
print('Accuracy on test data = ', metrics.accuracy_score(y_true = y_test, y_pred = y_pred))
#Make confusion matrix with the test dataset
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
