```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import scipy as sc
from scipy.stats import linregress
from scipy.stats import norm
import numpy as np
import warnings
warnings.filterwarnings("ignore")
#import contraceptives by method csv
method = pd.read_csv('By method.csv', encoding='latin-1',header = 0, index_col = False)
method.head(15)
```

```python
#import contraceptives by age and marital status csv
age_marital = pd.read_csv('By marital status and age_unedited.csv', encoding='latin-1',header = 0, index_col = False)
age_marital.head(15)
```

## 2. Clean Data

```python
#drop columns that will not be useful for our interpretation
method=method.iloc[0:1366,0:23]
#check info of dataset
method.info()
```

```python
#drop columns that will not be useful for our interpretation
age_marital=age_marital.drop(columns=['ISO code', 'DataCatalog ID', 'DataCatalog LongName', 'DataCatalog ShortName',
                         'Note on country', 'Note on data', 'Note on population', 'Unmet need:\nindicator'])
#check info of dataset
age_marital.info()
```

```python
#Use info to replace all NAs with 0
method.fillna(0)
#replace .. missing value with 0
method = method.replace(to_replace='..', value = 0)
#replace ... missing value with 0
method = method.replace(to_replace='...', value = 0)
#check if an NA values left in dataframe
print(method.isnull().values.any())
#refer to columns that need to be changed from object to float
cols = method.columns[5:22]
method[cols] = method[cols].apply(pd.to_numeric)
#conery survey years to int
method['Survey start year '] = method['Survey start year '] .astype(int)
method['Survey end year'] = method['Survey end year'] .astype(int)
age_marital['Survey\nstart year'] = age_marital['Survey\nstart year'] .astype(int)
age_marital['Survey\nend year'] = age_marital['Survey\nend year'] .astype(int)
#rename columns
method = method.rename({'ï»¿Country':'Country'}, axis=1)
age_marital = age_marital.rename({'Contraceptive use:\nmethods':'Contraceptive Methods'}, axis=1)
```

```python
#check overall data distribution and values are reasonable
print(method.describe())
print(age_marital.describe())
```
